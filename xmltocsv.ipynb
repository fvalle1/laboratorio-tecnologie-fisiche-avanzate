```python
%pylab inline
import pandas as pd
```

```python
def rowtoarray(row):
    output = []
    val=''
    for c in row:
        if c!=" ":
            val=val+c
        if c==" ":
            #print (val)
            output.append(float(val))
            val=''
    return output
```

```python
xmlfile = "Longrun.xml"
```

```python
file = open (xmlfile)
```

```python
distributions = {}
channels = ["0","1","2"]
for channel in channels:
    distributions[channel]=[]
```

```python
def appendToDist(row):
    if  row[:20]==' <trace channel="0">':
        #print (row[20:-1])
        #print("\n\n************")
        val = min(rowtoarray(row[20:-1]))
        if val !=0:
            distributions['0'].append(3333-val)
    if  row[:20]==' <trace channel="1">':
        val = min(rowtoarray(row[20:-1]))
        if val !=0:
            distributions['1'].append(4095-val)
    if  row[:20]==' <trace channel="2">':
        val = min(rowtoarray(row[20:-1]))
        if val !=0:
            distributions['2'].append(3641-val)
```

```python
file = open (xmlfile)
for row in file:
    appendToDist(row)
```

```python
for ch in channels:
    df = pd.DataFrame(distributions[ch])
    df.to_csv("dist"+ch+".csv", header=False, index=False)
```
